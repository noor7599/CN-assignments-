```python

words = ['My', 'name', 'is', 'Noor']
word_to_num = {'My': 0, 'name': 1, 'is': 2, 'Noor': 3}
num_to_word = {0: 'My', 1: 'name', 2: 'is', 3: 'Noor'}

X = [0, 1, 2] 
Y = 3          
vocab = 4
hidden = 4
lr = 0.1
import random
def rand(): return random.uniform(-0.1, 0.1)

Wxh = [[rand() for _ in range(vocab)] for _ in range(hidden)]   
Whh = [[rand() for _ in range(hidden)] for _ in range(hidden)]  
Why = [[rand() for _ in range(hidden)] for _ in range(vocab)]   
bh = [0]*hidden
by = [0]*vocab

def tanh(x): return (pow(2.718, x) - pow(2.718, -x)) / (pow(2.718, x) + pow(2.718, -x))
def dtanh(y): return 1 - y*y
def softmax(v):
    exps = [pow(2.718, i) for i in v]
    s = sum(exps)
    return [i/s for i in exps]
def onehot(i, size):
    v = [0]*size
    v[i] = 1
    return v

for epoch in range(100):
    h = { -1: [0]*hidden }
    for t in range(3): 
        x = onehot(X[t], vocab)
        h[t] = []
        for i in range(hidden):
            total = sum(Wxh[i][j]*x[j] for j in range(vocab)) + sum(Whh[i][j]*h[t-1][j] for j in range(hidden)) + bh[i]
            h[t].append(tanh(total))
    y = []
    for i in range(vocab):
        total = sum(Why[i][j]*h[2][j] for j in range(hidden)) + by[i]
        y.append(total)
    probs = softmax(y)

    dWhy = [[0]*hidden for _ in range(vocab)]
    dWxh = [[0]*vocab for _ in range(hidden)]
    dWhh = [[0]*hidden for _ in range(hidden)]
    dbh = [0]*hidden
    dby = [0]*vocab

    dy = probs[:]
    dy[Y] -= 1 

    for i in range(vocab):
        for j in range(hidden):
            dWhy[i][j] += dy[i] * h[2][j]
        dby[i] += dy[i]

    dh_next = [0]*hidden
    for t in reversed(range(3)):
        dh = [0]*hidden
        for i in range(hidden):
            for j in range(vocab):
                dh[i] += Why[j][i]*dy[j] if t == 2 else 0
            dh[i] += dh_next[i]
            dh_raw = dh[i] * dtanh(h[t][i])

            x = onehot(X[t], vocab)
            for j in range(vocab):
                dWxh[i][j] += dh_raw * x[j]
            for j in range(hidden):
                dWhh[i][j] += dh_raw * h[t-1][j]
            dbh[i] += dh_raw
        
        dh_next = [sum(Whh[j][i]*dh_raw for j in range(hidden)) for i in range(hidden)]

    for i in range(hidden):
        for j in range(vocab):
            Wxh[i][j] -= lr * dWxh[i][j]
    for i in range(hidden):
        for j in range(hidden):
            Whh[i][j] -= lr * dWhh[i][j]
    for i in range(vocab):
        for j in range(hidden):
            Why[i][j] -= lr * dWhy[i][j]
    for i in range(hidden):
        bh[i] -= lr * dbh[i]
    for i in range(vocab):
        by[i] -= lr * dby[i]

    if epoch % 10 == 0:
        pred = probs.index(max(probs))
        print(f"Epoch {epoch}: Predicted = {num_to_word[pred]}, Target = Noor")
```
